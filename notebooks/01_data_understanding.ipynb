## Data Understanding 

```python
import pandas as pd
```

#### Load all datasets

```python
begin_inventory = pd.read_csv("../data/raw/begin_inventory.csv")
end_inventory = pd.read_csv("../data/raw/end_inventory.csv")
purchase_prices = pd.read_csv("../data/raw/purchase_prices.csv")
purchases = pd.read_csv("../data/raw/purchases.csv")
sales = pd.read_csv("../data/raw/sales.csv")
vendor_invoice = pd.read_csv("../data/raw/vendor_invoice.csv")
vendor_sales_summary = pd.read_csv("../data/raw/vendor_sales_summary.csv")
```

data set summary

```python
def dataset_summary(df,name):
    print("="*60)
    print(f"Dataset : {name}")
    print("="*60)

    print("\n Shape:")
    print(df.shape)

    print("\nColumns:")
    print(df.columns.tolist())

    print("\nInformation:")
    print(df.info())

    print("\n Data Types:")
    print(df.dtypes)

    print("\nMissing Values:")
    print(df.isnull().sum())

    print("\nDuplicate Rows:")
    print(df.duplicated().sum())

    print("\nFirst 5 rows:")
    print(df.head())
```

Vendor Summary

```python
dataset_summary(vendor_invoice, "Vendor Invoice")
```

Begin Inventory

```python
dataset_summary(begin_inventory, "Begin Inventory")
```

End Inventory

```python
dataset_summary(end_inventory, "End Inventory")
```

Purchase Price

```python
dataset_summary(purchase_prices, "Purchase Prices")
```

Purchases

```python
dataset_summary(purchases, "Purchases")
```

Sales

```python
dataset_summary(sales, "Sales")
```

Vendor Sales Summary

```python
dataset_summary(vendor_sales_summary, "Vendor Sales Summary")
```
